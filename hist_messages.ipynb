```python
from collections import Counter

import numpy as np
import matplotlib.pyplot as plt

```

```python
flatten = lambda list_: [item for sublist in list_ for item in sublist]
```

```python
epochs = [10, 100, 200, 300, 400, 500]
plt.figure(figsize=(12,6))
for epoch in epochs:
    file_ = np.load(f'messages/messages_{epoch}.npy', allow_pickle=True)
    l = flatten([list(file_[i]) for i in range(100)])
    c = Counter(l)
    values_sorted = sorted(c.values(), reverse=True)
    values_sorted = [value/len(l) for value in values_sorted]
    plt.bar(range(len(values_sorted)), values_sorted, label=f'Epoch {epoch}')
plt.legend();
```
